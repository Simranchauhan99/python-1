```python
#line chart
from matplotlib import pyplot as plt
#plotting to our canvas 
plt.plot([1,2,3],[4,5,7])
#showing what we plot
plt.show()
```

```python
#adding title 
x= [5,8,10]
y= [12, 16 , 6]
plt.plot(x,y)
plt.title('info')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.show()
```

```python
#adding colour
from matplotlib import style
style.use('ggplot')
x = [5, 8, 10]
y = [12, 16, 6]
x2 = [6, 9, 11]
y2 = [6, 15, 17]
plt.plot(x ,y ,'g')
plt.plot(x2,y2,'c')
plt.title('epic info')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.legend()
plt.grid(true, color = 'k')
plt.show()
```

```python
#bar graph
plt.bar([1,3,5,7,9],[5,2,7,8,2], label= "ex 1")
plt.bar([2,4,6,8,10],[8,6,2,5,6], label= "ex 2" , color = 'g')
plt.legend
plt.xlabel('bar num')
plt.ylabel('bar height')
plt.title('info')
plt.show
    
```

```python
#histogram
population_ages =[22,55,62,45,21,22,32,42,42,49,69]
bins=[0,10,20,30,40,50,60]
plt.hist(population_ages,bins )
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('histogram')
plt.show()
```

```python
#scatter
x= [1,2,3,4,5,6,7,8]
y= [5,2,4,2,1,4,5,2]
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('scatter plot')
plt.legend()
plt.show()
```

```python
#stackplot
days = [1,2,3,4,5,6]
sleeping= [7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]
plt.plot([],[], color='m')
plt.plot([],[], color='c')
plt.plot([],[], color='r')
plt.plot([],[], color='k')
plt.stackplot(days,sleepling,eating,playing,working,colors=['m','c','r','k'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('stack plot')
plt.legend()
plt.show()
```
